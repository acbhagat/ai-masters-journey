```python
def print_full_ascii_table():
    print(f"{'Decimal':<10} {'Hex':<10} {'Character':<10}")
    print("-" * 35)
    for i in range(128):
        # Some control characters (0-31 and 127) are non-printable
        # and may appear as blank or cause formatting issues in some terminals.
        if 0 <= i <= 31 or i == 127:
            char_repr = repr(chr(i)) # Use repr() to show escape sequences like '\\n'
        else:
            char_repr = chr(i)
        print(f"{i:<10} {hex(i):<10} {char_repr:<10}")

print_full_ascii_table()
```

# Data Slicing

Positive Indexation = Forward Direction    

For printing elements in forward direction, syntax is

object_name[positive index of left-most element,positive index of right-most element]

----------------------------------------------------------

Negative Indexation = Forward Direction   

object_name[negative index of left-most element, negative index of right-most element]

----------------------------------------------------------

Positive Indexation = Reverse    

[positive index of right-most element : positive index of left-most element : -1]

----------------------------------------------------------

Negative Indexation = Reverse    

[negative index of right-most element : negative index of left-most element : -1]

```python
print(list(range(10)))
print(list(range(2, 12)))
```

```python
#Python uses 0 indexation
a = 'abcdefghijklmnopqrstuvwxyz'
print('The total number of characters in the string "a" is',len(a))
print('The index of "a" is',a.index('a'))
print('The index of "z" is',a.index('z'))
print('Positive Index of t is',a.index('t'))
print('Negative Index of t is',a.index('t') - len(a))
```

```python
#Forward direction with positive index
                       #---->
"""
We want to print characters from t to x
Positive index of t is 19
Positive index of x is 23
So if we want to go from character 't' to character 'x',
we will have to set the range from index of character 't' i.e. 19, to index of character 'y' i.e. 24
"""

a = 'abcdefghijklmnopqrstuvwxyz'
print('Positive Index of t is',a.index('t'))
print('Positive Index of x is',a.index('x'))
#a[19:23] : Index 19 to index 23 will go only till index 22
print('a[19:23] :' , a[19:23])
#a[19:24] : Index 19 to index 24 will go till index 23
print('a[19:24] :', a[19:24])
```

```python
#Forward direction with negative index
                       #---->


"""
We want to print characters from t to x
Negative index of t is -7
Negative index of x is -3
For printing character at index -3, we must give a prior index i.e. -2
So if we want to go until character 'x',
we will have to set the range from index of character 't' i.e. -7, to index of character 'y' i.e. -2
"""
a = 'abcdefghijklmnopqrstuvwxyz'
print('Negative Index of t is',a.index('t') - len(a))
print('Negative Index of x is',a.index('x') - len(a))
#a[-7:-3] : Index -7 to index -3 will go only till index -4
print('a[-7:-3]', a[-7:-3])
#a[-7:-2] : Index -7 to index -2 will go till index -3
print('a[-7:-2]', a[-7:-2])
```

```python
#Reverse direction with negative index
                      #<-----

"""
We want to print characters from x to t backwards, i.e. 'xwvut'
Negative index of x is -3
Negative index of t is -7
To start from the character 'x' , the start index is -3,
If we want to go until character 't', i.e. -7, we must give an index 1 later than t i.e. -8
We will have to set the range from index of character 'x' i.e. -3, to index of character 's' i.e. -8
Also since we wish to go reverse, we shall put a third parameter, i.e. -1
"""
a = 'abcdefghijklmnopqrstuvwxyz'

print('Negative Index of t',a.index('t') - len(a))
print('Negative Index of x',a.index('x') - len(a))
print('a[-7:-2:-1]',a[-7:-2:-1])
print('a[-3:-7:-1]',a[-3:-7:-1])
print('a[-3:-8:-1]',a[-3:-8:-1])
```

```python
#Reverse direction with positive index

#Starting from 'k' to 'd', we want to print the letters in reverse i.e. kjihgfed

"""
Positive index of 'k' is 10
Positive index of 'd' is 3
To start from the character 'k' , the start index is 10,
If we want to go until character 'd', i.e. 3, we must give index of 'd' minus 1 i.e. 2
We will have to set the range from index of character 'k' i.e. 10, to index of character 'c' i.e. 2
Also since we wish to go reverse, we shall put a third parameter, i.e. -1
"""
a = 'abcdefghijklmnopqrstuvwxyz'
b = list(a)
print(b)
print('Positive Index of k',b.index('k'))
#print('Negative Index of t',b.index('k') - len(b))
print('Positive Index of d',b.index('d'))
#print('Negative Index of x',b.index('d') - len(b))
print('a[10:3:-1]', a[10:3:-1])
print('a[10:2:-1]',a[10:2:-1])
```

# Set

- An unordered sequence of unique elements.
- Set does not have any indexing.
- We cannot refer/call any element of a set by an index.
- Set does not accept duplicated values.
- Sets are mutable, which means we can modify a set by adding, removing, replacing elements.
- To create a set, we use curly/flower brackets.


|Method|Description|
|------|-----------|
|add()|Adds an element to the set|
|clear()|Removes all the elements from the set|
|copy()|Returns a copy of the set|
|difference()|Returns a set containing the difference between two or more sets|
|difference_update()|Removes the items in this set that are also included in another, specified set|
|discard()|Remove the specified item|
|intersection()|(Commutative) Returns a set, that is the intersection of two or more sets|
|intersection_update()|Removes the items in this set that are not present in other, specified set(s)|
|isdisjoint()|(Commutative) Returns whether two sets have a intersection or not|
|issubset()|Returns whether another set contains this set or not|
|issuperset()|Returns whether this set contains another set or not|
|pop()|Removes an element from the set|
|remove()|Removes the specified element|
|symmetric_difference()|(Commutative) Returns a set with the symmetric differences of two sets|
|symmetric_difference_update()|inserts the symmetric differences from this set and another|
|union()|(Commutative) Return a set containing the union of sets|
|update()|Update the set with another set, or any other iterable|

```python
set1 = {'Apple', 'Mango', 'Banana', 'Orange', 'Pineapple'}
set2 = {'Banana', 'Chickoo','Orange','Grapes','Papapa'}
set3 = {'Mango', 'Dragon Fruit', 'Muskmelon', 'Watermelon', 'Orange', 'Grapes'}
print('set1^set2', set1.symmetric_difference(set2))
print('set3',set3)
print('set1^set2^set3',set1.symmetric_difference(set2).symmetric_difference(set3))
```

Symmetric Difference will always consider only 2 sets at a time.

For example, we have 5 sets : Set1, Set2, Set3, Set4, Set5

So this is how it will calculate response for the command below:

Set1^Set2^Set3^Set4^Set5

- First it will find symm diff of set1 & set2 only. Save this response in a.
- In second iteration it will do symm diff a & set3, save this response in b.
- In third iteration, it will symm diff b & set4, save this response in c.
- In final iteration, it will symm diff c & set5.
That will be final response.

```python
set1 = {'Apple', 'Mango', 'Banana', 'Orange', 'Pineapple'}
set2 = {'Banana', 'Chickoo','Orange','Grapes','Papapa','Pineapple'}
set3 = {'Mango', 'Dragon Fruit', 'Muskmelon', 'Watermelon', 'Orange', 'Grapes','Pineapple'}
set4 = {'Pineapple','Orange', 'Mango', 'Muskmelon', 'Apple'}
set5 = {'Blueberry', 'Strawberry', 'Pineapple','Orange', 'Papaya'}
print(set1^set2^set3^set4^set5)
print('\n')
print('Union of all 5 sets')
print(set1 | set2 | set3 | set4 | set5)
print('Intersection of all 5 sets')
print(set1 & set2 & set3 & set4 & set5)
print('Union of all 5 sets minus Intersection of all 5 sets')
print((set1 | set2 | set3 | set4 | set5) - (set1 & set2 & set3 & set4 & set5))
```
